```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://www.applitrack.com/mnps/onlineapp/jobpostings/Output.asp?all=1&'
resp = requests.get(url)
html = BeautifulSoup(resp.content, 'html.parser')
```

```python
df = pd.DataFrame()

for posting in html.select("ul"):
    if posting.get('class') == ["\\'postingsList\\'"]:
        job = dict()
        title_list = posting.select('li')[0]
        title, pos_id_span = list(title_list.children)
        job['title'] = title
        pos_id = pos_id_span.get_text().replace('JobID:', '').strip()

        spans = posting.select('span.normal')
        job['date_posted'] = spans[0].get_text()
        job['department'] = spans[1].get_text()
        job['additional_info'] = spans[-1].get_text()
        
        df = df.append(pd.Series(job, name=pos_id))
```
